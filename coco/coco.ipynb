```python
import os
import zipfile

import requests
from tqdm import tqdm

data_type = "trainval2017"
instance_type = "val2017"
dataset_url = f"http://images.cocodataset.org/annotations/annotations_{data_type}.zip"


def download_to(path: str, url: str):
    """Download a file from the given URL to the specified path.

    Args:
        path (str): The path to save the downloaded file.
        url (str): The URL to download the file from.
    """
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get("content-length", 0))
    chunk_size = 1024
    print(f"Downloading to {path}")

    with (
        open(path, "wb") as file,
        tqdm(
            desc="Downloading",
            total=total_size,
            unit="B",
            unit_scale=True,
            unit_divisor=chunk_size,
        ) as bar,
    ):
        for data in response.iter_content(chunk_size):
            file.write(data)
            bar.update(len(data))


download_to("annotations.zip", dataset_url)

# Unzip the annotations file
with zipfile.ZipFile("annotations.zip", "r") as zip_ref:
    zip_ref.extractall(os.getcwd())
```

```python
from pycocotools.coco import COCO

annFile = f"{os.getcwd()}/annotations/instances_{instance_type}.json"
print(f"Loading annotation file {annFile}")

coco = COCO(annFile)
```

```python
category_ids = coco.getCatIds()
num_categories = len(category_ids)
print(f"Number of categories: {num_categories}")
for ids in category_ids:
    cats = coco.loadCats(ids)
    print(cats)
```

```python
categories = coco.loadCats(category_ids)
cat_names = [cat["name"] for cat in categories]
print(f"Categories: {cat_names}")
with open("coco.names", "w") as f:
    for item in cat_names:
        f.write("%s\n" % item)

# Note coco.names-fr was translated from coco.names using ChatGPT
```
